```python
%matplotlib inline
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import math 
import seaborn as sns
```

```python
# read in raw csv data file
raw_df_movie = pd.read_csv('/Users/gavinfosterhome/Metis/projects/Luther/movie_data.csv', parse_dates={'Date':['ReleaseDate']})
raw_df_movie['Date'] = raw_df_movie['Date'].astype(datetime)

#nans = df_movie.shape[0] - df_movie.dropna().shape[0]
#print('%d rows have missing values' % nans)
#df_movie[df_movie['Directors'].isnull()]
#df_movie[df_movie['Directors'].notnull()]
# df_movie.fillna(value=0, inplace=True)
# df_movie.head(50)

# process raw data
df_movie = raw_df_movie.dropna(subset=['DomesticTotalGross','ProductionBudget','Genre','RuntimeInMinutes','MPAARating','OpeningWeekendGross','OWNumberTheaters'])
df_movie['Ones'] = 1.0
df_movie = df_movie.rename(columns = {'DomesticTotalGross':'DomesticGross'})
df_movie = df_movie.rename(columns = {'ProductionBudget':'Budget'})
df_movie = df_movie.rename(columns = {'RuntimeInMinutes':'Runtime'})
df_movie = df_movie.rename(columns = {'MPAARating':'Rating'})
df_movie = df_movie.rename(columns = {'OpeningWeekendGross':'OWeekend'})
df_movie = df_movie.rename(columns = {'OWNumberTheaters':'OWTheaters'})
df_movie = df_movie.rename(columns = {'OWAvgGrossPerTheater':'OWAvg'})
#df_movie['StarRating'] = pd.Series('', index=df_movie.index) # or df_movie.insert(loc=8, column='position', value='')

# df_movie = df_movie.log('DomesticGross')
df_movie['LogGross']= np.log(df_movie.DomesticGross)
df_movie['LogBudget']= np.log(df_movie.Budget)
df_movie['LogOWT']= np.log(df_movie.OWTheaters)
```

# Test for Production Budget as the single feature

```python
import statsmodels.formula.api as sm

X = df_movie[['Budget', 'Ones']]
#X = df_movie[['LogOWT', 'Ones']]
Y = df_movie['DomesticGross']
#Y = df_movie['LogGross']

sm_linmodel = sm.OLS(Y, X).fit()
sm_linmodel.summary()
```

```python
my_plot = df_movie.sort(columns='Date',ascending=False).plot('Budget','DomesticGross',title="Total Domestic Gross over Budget",legend=None, fontsize=10, figsize=(14, 8))
my_plot.set_ylabel("Domestic Gross")
my_plot.set_xlabel("Budget")
```

```python
# Plot a graph on train data set 1
plt.scatter(df_movie.Budget, df_movie.DomesticGross)
plt.title('Box Office Gross Revenue vs Production Budget')
plt.xlabel('Budget')
plt.ylabel('Box Office')
plt.plot(df_movie.Budget, sm_linmodel.predict(), 'rs')
```

# Set two cross-validation data sets

```python
#Cross Validation data sets
train_data1 = df_movie.ix[:11650] # approximately 75% but not exact
test_data1 = df_movie.ix[11651:]  # approximately 25% but not exact

train_data2 = df_movie.ix[:5550] # approximately 33% but not exact
test_data2 = df_movie.ix[551:]  # approximately 66% but not exact
#train_data1['Genre'].apply(str)
```

# Test a two feature model using Production Budget and Opening Week Theaters

```python
# Running Regression Model using training data set 1
import statsmodels.formula.api as trsm1

X_tr1 = train_data1[['LogBudget','LogOWT','Ones']]
#X_tr1 = pd.concat([X_tr1, pd.get_dummies(train_data1['Genre'])], axis=1)
print X_tr1.head()
Y_tr1 = train_data1['LogGross']
trd1_linmodel = trsm1.OLS(Y_tr1, X_tr1).fit()
trd1_linmodel.summary()
```

```python
# Plot a graph on train data set 1
plt.scatter(train_data1.LogBudget, train_data1.LogGross)
plt.title('Scatter plot: Box Office Domestic Gross vs Budget')
plt.xlabel('Log Budget')
plt.ylabel('Log Box Office')
plt.plot(train_data1.LogBudget, trd1_linmodel.predict(), 'rs')
```

```python
# Plot a graph on train data set 1 - Opening Week Theaters
plt.scatter(train_data1.OWTheaters, train_data1.LogGross)
plt.title('Scatter plot: Box Office Domestic Gross vs Budget')
plt.xlabel('OWTheaters')
plt.ylabel('Log Box Office')
plt.plot(train_data1.OWTheaters, trd1_linmodel.predict(), 'rs')
```

```python
# Plot a graph on train data set 1-  Runtime
plt.scatter(train_data1.Runtime, train_data1.LogGross)
plt.title('Scatter plot: Box Office Domestic Gross vs Budget')
plt.xlabel('Runtime')
plt.ylabel('Log Box Office')
plt.plot(train_data1.Runtime, trd1_linmodel.predict(), 'rs')
```

```python
#Comparing predicted vs actual in test data 1
X_ts1 = test_data1[['LogBudget', 'LogOWT','Ones']]
Y_ts1 = test_data1['LogGross']
Y_pred_ts1 = trd1_linmodel.predict(X_ts1)
plt.plot(Y_ts1, Y_pred_ts1,'go')
plt.title('Comparison of Predicted vs Actual Box Office (Using a Log Scale)')
plt.xlabel('Actual Gross')
plt.ylabel('Predicted Gross by Model')

# 1:1 line
plt.plot(Y_ts1, Y_ts1, 'm--', label = 'Ideal Prediction')
plt.legend()
```

```python
#Cross Validation - train set 2

# Running Regression Model using training data set 2
import statsmodels.formula.api as trsm2

X_tr2 = train_data2[['LogBudget','LogOWT','Ones']]
Y_tr2 = train_data2['LogGross']
trd2_linmodel = trsm2.OLS(Y_tr2, X_tr2).fit()
trd2_linmodel.summary()
```

```python
# Plot a graph on train data set 2
plt.scatter(train_data2.LogBudget, train_data2.LogGross)
plt.xlabel('Log Budget')
plt.ylabel('Log Box Office')
plt.plot(train_data2.LogBudget, trd2_linmodel.predict(), 'rs')
```

```python
# Plot a graph on train data set 2
plt.scatter(train_data2.LogOWT, train_data2.LogGross)
plt.xlabel('LogOWT')
plt.ylabel('Log Box Office')
plt.plot(train_data2.LogOWT, trd2_linmodel.predict(), 'rs')
```

```python
#Comparing predicted vs actual in test data 2
X_ts2 = test_data2[['LogBudget', 'LogOWT','Ones']]
Y_ts2 = test_data2['LogGross']
Y_pred_ts2 = trd2_linmodel.predict(X_ts2)
plt.plot(Y_ts2, Y_pred_ts2,'go')
plt.title('Comparison of Predicted vs Actual Box Office')
plt.xlabel('Actual Gross')
plt.ylabel('Predicted Gross by Model')

# 1:1 line
plt.plot(Y_ts2, Y_ts2, 'm--', label = 'Ideal Prediction')
plt.legend()
```
